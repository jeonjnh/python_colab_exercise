# 11. 데이터 정렬

```python
import pandas as pd
```

```python
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

df = pd.DataFrame(data, index=['No.1','No.2','No.3','No.4','No.5','No.6','No.7','No.8'])
df.index.name = '지원번호'
df
```

```python
df.sort_values('키')
```

```python
df.sort_values('키', ascending=False)
```

```python
df.sort_values(['수학', '영어'], ascending=False)
```

```python
df.sort_values(['수학', '영어'], ascending=True)
```

```python
df.sort_values(['수학', '영어'], ascending=[True, False])
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고', '능남고':'무슨고'})
```

```python
df
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace=True)#
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교'
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['과학'] + df['사회'] + df['영어'] + df['수학']
df
```

```python
df['결과'] = 'Fail'
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass'
df
```
