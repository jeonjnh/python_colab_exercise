# 8. 데이터 선택 (iloc)

```python
import pandas as pd
```

```python
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

df = pd.DataFrame(data, index=['No.1','No.2','No.3','No.4','No.5','No.6','No.7','No.8'])
df.index.name = '지원번호'
df
```

```python
df.iloc[0] # index 0 data
```

```python
df.iloc[4]
```

```python
df.iloc[0:5] # index 0 - 4
```

```python
df.iloc[0,1] 
```

```python
df.iloc[4, 2] #강백호 키
```

```python
df.iloc[[0,1], 2]
```

```python
df.iloc[[0,1], [3,4]]
```

```python
df.iloc[0:5, 3:8]
# df.iloc[ row, column]
```

# 09. 데이터 선택(조건)

```python
df
```

```python
df['키'] >= 185 # 학생들의 키가 185이상인지 여부를 boolian 
```

```python
filter_height = (df['키'] >= 185)
df[filter_height]
```

```python
df[~filter_height]
```

```python
df[df['키']>=185]
```

```python
# df.loc[row_sel, col_sel]
df.loc[df['키']>=185, '수학']
```

```python
df.loc[df['키']>=185, ['수학', '과학']]
```

```python
# & : and
df.loc[(df['키']>=185) & (df['학교'] == '북산고')]
```

```python
# | : or
df.loc[(df['키'] < 170)|(df['키'] > 200)]
```

## String 

```python
filt = df['이름'].str.startswith('송')
df[filt]
```

```python
df[df['이름'].str.startswith('송')]
```

```python
df[df['이름'].str.contains('태')]
```

```python
df[~filt]
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs)
df[filt]
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java')
df[filt]
```
